## DFSアルゴリズムを使った探索

```python
import numpy as np
from graphviz import Graph, Digraph
```

```python
##  既知情報  ##
node_num = 15
edge_num = 14

edges = [(0, 1), (1, 2), (1, 3),
                 (0, 4), (4, 5), (5, 6), (5, 7), (4, 8), (8, 9), (8, 10),
                 (0, 11), (11, 12), (11, 13), (13, 14)]
np_edges = np.array(edges)

## 探索条件
con_start = 0
con_goal = 14
```

```python
## グラフ構造のヴィジュアライズ
graph = Digraph(format='png')

## ノードの作成
for i in range(node_num):
    if i == con_start:
        graph.attr("node", color="green", fontcolor="green")
    elif i == con_goal:
        graph.attr("node", color="red", fontcolor="red")
    else:
        graph.attr("node", color="black", fontcolor="black")
    graph.node(str(i))

## エッジの作成
for i, j in edges:
    graph.edge(str(i), str(j))

graph.render("tree")
display(graph)
```

```python
class Node:
    def __init__(self, to, weight=1):
        self.to = to
        self.weight = weight
```

```python
#ノードリストの作成
nodes = []
for i in range(node_num):
    matched_indexes = np.where(np_edges[:,0]==i)
    element = np_edges[matched_indexes][:,1]
    
    nd = Node(element)
    nodes.append(nd)
nodes
```

```python
## 初期設定
stack = []
start_instance = nodes[0]
ikisaki = start_instance.to
list_ikisaki = ikisaki.tolist()
stack.extend(list_ikisaki)
```

```python
## 初期設定
stack = []
stack.extend(nodes[con_start].to.tolist())
i = 0
reachable = False

## 探索
print("Seq : Node")
while(len(stack)!=0):
    poped_num = stack.pop()
    stack.extend(nodes[poped_num].to.tolist())
    i += 1
    print("{0}  :  {1}".format(i, poped_num))
    if poped_num == con_goal:
        reachable = True

print("-"*20)
print("{0} -> {1} : {2}".format(con_start, con_goal, reachable))
```
